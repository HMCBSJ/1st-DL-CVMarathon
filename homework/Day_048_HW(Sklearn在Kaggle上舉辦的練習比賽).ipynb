```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#讀入資料
path = 'data/data-science-london-scikit-learn/'

#header為加上欄位標號
x_train_all = pd.read_csv(path + 'train.csv', header = None)
y_train_all = pd.read_csv(path + 'trainLabels.csv', header = None)
x_test = pd.read_csv(path + 'test.csv', header = None)
```

```python
x_train_all
```

```python
#資料切分成train跟validation data
from sklearn.model_selection import train_test_split

x_train,x_val,y_train,y_val = train_test_split(x_train_all, y_train_all, test_size = 0.2, random_state = 42)
x_train
```

### RandomForest

```python
from sklearn import datasets, metrics
from sklearn.model_selection import KFold, GridSearchCV

from sklearn.ensemble import RandomForestClassifier

#嘗試使用Random Forest
clf_rf = RandomForestClassifier(n_estimators=300, criterion='gini', max_depth=10 ,random_state=42)
clf_rf.fit(x_train, y_train)
```

```python
y_val_pred = clf_rf.predict(x_val)
print(metrics.accuracy_score(y_val, y_val_pred))
```

```python
#使用 hyper-parameter search 在random forest找出最佳參數

n_estimators = [100, 200, 300, 400, 500]
max_depth = [5, 10, 15, 20]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
grid_search = GridSearchCV(clf_rf, param_grid, scoring='accuracy', n_jobs=-1, verbose=1)

# 開始搜尋最佳參數
grid_result = grid_search.fit(x_train, y_train)

# 預設會跑 5-fold cross-validadtion，總共 20 種參數組合，總共要 train 100 次模型
```

```python
# 印出最佳結果與最佳參數
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print(grid_result.best_params_)
```

```python
clf_rf_best = RandomForestClassifier(n_estimators=400, criterion='gini', max_depth=15 ,random_state=42)
clf_rf_best.fit(x_train, y_train)
y_val_pred = clf_rf_best.predict(x_val)
print(metrics.accuracy_score(y_val, y_val_pred))
```

### 嘗試高斯混合模型(此比賽的大神所使用，理解與筆記)

筆記：

    高斯混和模型是一種聚類演算法，假設每個資料皆由一個高斯分布所產生，而這一批資料共同由幾個高斯分布產生，將所有來自不同分布的資料點混和在一起，因此稱為高斯混合模型

```python
x_train_all
```

```python
#參考https://github.com/siddharthagarwal/Kaggle-Data-Science-London-Scikit-Learn

from sklearn.mixture import GaussianMixture

#作用類似於pd.concat()，此用於拼接兩個矩陣，np.c_左右接和np.r_上下接
x_all = np.r_[x_train_all,x_test] 
lowest_bic = np.infty
bic = []
cv_types = ['spherical', 'tied', 'diag', 'full']
#類似grid_search找最佳參數
for cv_type in cv_types:
    for n_components in range(1, 7):
        gmm = GaussianMixture(n_components=n_components, covariance_type=cv_type)
        gmm.fit(x_all)
        
        #aic為可以計算數據擬合的程度，並避免過擬合，值越小表示模型越優良
        bic.append(gmm.aic(x_all))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm
            
best_gmm.fit(x_all)
x_train_all_gmm = best_gmm.predict_proba(x_train_all)
x_test_gmm = best_gmm.predict_proba(x_test)
```

```python
svc_gmm = SVC(gamma = 'auto', C = 1)
svc_gmm.fit(x_train_all_gmm,y_train_all.values.ravel())
y_pred = svc_gmm.predict(x_train_all_gmm)
print(metrics.accuracy_score(y_train_all, y_pred))
```

```python
C  = [1, 2, 3, 4, 5]
param_grid = dict(C = C)

## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
grid_search = GridSearchCV(svc_gmm, param_grid, scoring='accuracy', n_jobs=-1, verbose=1)

# 開始搜尋最佳參數
grid_result = grid_search.fit(x_train_all, y_train_all)

# 印出最佳結果與最佳參數
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print(grid_result.best_params_)
```

```python
svc_gmm_best = SVC(gamma = 'auto', C = 3, random_state=42)
svc_gmm_best.fit(x_train_all_gmm,y_train_all.values.ravel())
y_pred = svc_gmm_best.predict(x_train_all_gmm)
print(metrics.accuracy_score(y_train_all, y_pred))
```
